```python
# “header” include statements
import numpy as np
import math
from numpy import sqrt
import matplotlib.pyplot as plt
import scipy.constants as cnst

c = cnst.c
pi = cnst.pi
```

```python
# set parameters of program
wavelength = 780e-9 #m
N = 5
L = 27e-3 #m
n2 = 1.823
```

```python
# define a frequency range
f0 = int (c / wavelength)
delta_f = int(10e9)
numsteps_f = 10000
step_f = int(2*delta_f/ numsteps_f)
f_list = range(f0 - delta_f, f0 + delta_f, step_f)
```

```python
# define a function to compute transmission at a given frequency
# Use only the real part of the wave, and square to put in units of power.
def Transmission_ing(freq):
    E_t_img = 0
    reference_phase = np.exp(-1j*freq/c * L*n2*2*pi)

    for i in range(1, N+1):
        E_t_img += (np.exp(1j*(i-1)*2*pi) * t1*t2 * (r1*r2)**(i-1)
                    * np.exp(1j*freq/c * (2*(i-1)+1)*L*n2*2*pi)
                    * reference_phase)

    return np.abs(E_t_img)**2
```

```python
# compute electric field at each frequency
R_values = [0.1, 0.3, 0.8]
transmitted_intensities_dict = {}

for R in R_values:
    r1 = np.sqrt(R)
    r2 = r1
    t1 = np.sqrt(1 - r1**2)
    t2 = t1

    transmitted_intensities = [Transmission_ing(f) for f in f_list]
    transmitted_intensities_dict[R] = transmitted_intensities
```

```python
# Plot
plt.figure(figsize=(10,6))

for R in R_values:
    #FSR
    FSR = c/2*L
    # Finesse
    F = (np.pi * np.sqrt(R)) / (1 - R)

    normalized_intensities = transmitted_intensities_dict[R] / np.max(transmitted_intensities_dict[R])
    plt.plot((np.array(f_list) - f0)/1e9, normalized_intensities, label=f"R={R}, FSR={FSR/1e6:.2} MHz, F={F:.2f}")

plt.title("Transmitted Power vs Frequency for Different R")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Transmitted Power (W)")
plt.legend()
plt.grid(True)
plt.ylim([0, 1.1])
plt.tight_layout()
plt.show()

print(f'FSR={FSR/1e6:.2} MHz, F={F:.2f}')
```
