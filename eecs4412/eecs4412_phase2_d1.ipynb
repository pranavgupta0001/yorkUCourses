```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
```

```python

# Set directory path
dir_path = '/content/yorkUCourses/eecs4412/data'

# Get list of files in directory
files = os.listdir(dir_path)

# Load data from each file into a dictionary
data = {}
for i, file in enumerate(files):
    if file.endswith('.csv'):
        # Load data from CSV file
        df = pd.read_csv(os.path.join(dir_path, file))

        # Add filename as a column in the DataFrame
        df['filename'] = file

        # Add DataFrame to dictionary with company number as key
        data[i+1] = df


```

```python
for key in data:
  # Plot graphs
  plt.plot(data[key]['Date'], data[key]['Open'], label='Open')
  plt.plot(data[key]['Date'], data[key]['High'], label='High')
  plt.plot(data[key]['Date'], data[key]['Low'], label='Low')
  plt.plot(data[key]['Date'], data[key]['Close'], label='Close')
  plt.plot(data[key]['Date'], data[key]['Adj Close'], label='Adj Close Stock Price')
  plt.plot(data[key]['Date'], data[key]['Volume'], label='Volume')

  # Add title and labels to the graph
  plt.title('Stock Prices')
  plt.xlabel('Date')
  plt.ylabel('Price')

  # Add legend to the graph
  plt.legend()

  # Show the graph
  plt.show()
  break
```

```python
for key in data:
  # Calculate average close price and total volume
  avg_close_price = data[key]['Close'].mean()
  total_volume = data[key]['Volume'].sum()

  # Plot both values in the same chart
  fig, ax1 = plt.subplots()

  color = 'tab:red'
  ax1.set_xlabel('Date')
  ax1.set_ylabel('Average Close Price', color=color)
  ax1.plot(data[key]['Date'], data[key]['Close'], color=color)
  ax1.tick_params(axis='y', labelcolor=color)

  ax2 = ax1.twinx()

  color = 'tab:blue'
  ax2.set_ylabel('Total Volume', color=color)
  ax2.plot(data[key]['Date'], data[key]['Volume'], color=color)
  ax2.tick_params(axis='y', labelcolor=color)
  ax2.legend()
  fig.tight_layout()
  plt.show()
  break
```

```python
# Assume 'combined_df' is the combined dataframe containing the stock price data

# Apply Min-Max scaling on 'Close' dimension
combined_df = data[4]
combined_df['Close_scaled'] = (combined_df['Close'] - combined_df['Close'].min()) / (combined_df['Close'].max() - combined_df['Close'].min())

# Apply log transformation on 'Volume' dimension
combined_df['Volume_log'] = np.log(combined_df['Volume'])

# Create a scatter plot to compare the before and after data
plt.figure(figsize=(7, 6))
plt.scatter(combined_df['Volume'],combined_df['Close'], label='Before', color='red')
plt.scatter(combined_df['Volume_log'], combined_df['Close_scaled'], label='After', color='black')
plt.ylabel('Close')
plt.xlabel('Volume')
plt.title('Preprocessing Techniques Effect')
plt.legend()
plt.show()
plt.scatter(combined_df['Close_scaled'], combined_df['Volume_log'], label='After Preprocessing')
```
